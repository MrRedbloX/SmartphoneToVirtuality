```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

U = np.array([0, 1, 2])
V = np.array([1, 0, 0])
W = np.array([-0.5, -2, -1])
X = np.array([0, 0, 0])

camera1 = [
    np.array([0.10931544, 0.99395072, 0.01058803]), #A -> Z
    np.array([0.39272138, -0.43277323, 0.81146612]), #Z -> E
    np.array([ 0.01103039, -0.05724966, -0.99829896]) #E -> A
]

camera2 = [
    np.array([0.39272138, -0.43277323, 0.81146612]), #Z -> E
    np.array([ 0.01103039, -0.05724966, -0.99829896]), #E -> A
    np.array([0.10931544, 0.99395072, 0.01058803]), #A -> Z
]

camera3 = [
    np.array([ 0.01103039, -0.05724966, -0.99829896]), #E -> A
    np.array([0.10931544, 0.99395072, 0.01058803]), #A -> Z
    np.array([0.39272138, -0.43277323, 0.81146612]), #Z -> E
]

# C2 = X + camera1
# C1 = C2 + camera2
# print(C2)
# print(C1)
# print(np.linalg.norm(-P2))

def get_error_distance(u, v, w, O):
    A = O
    B = A + u
    C = B + v
    
    Cp = C + w
    
    print(Cp)

    dist = np.linalg.norm(A-Cp)
    print(dist)

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot([A[0], B[0]], [A[1], B[1]], zs=[A[2], B[2]], label='1')
    ax.plot([B[0], C[0]], [B[1], C[1]], zs=[B[2], C[2]], label='2')
    ax.plot([C[0], Cp[0]], [C[1], Cp[1]], zs=[C[2], Cp[2]], label='3')
    ax.legend()
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    plt.show()
    
get_error_distance(camera1[0], camera1[1], camera1[2], X)
get_error_distance(camera2[0], camera2[1], camera2[2], X)
get_error_distance(camera3[0], camera3[1], camera3[2], X)
```
